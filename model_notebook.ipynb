#### Description of Data:
1. Age : age of the person
2. Sex : gender(1 = Male, 0=Female)
3. cp : Chest Pain Type (0 = Typical angina, 1 = atypical angina, 2 = non-anginal pain, 3 = asymptomatic)
4. testbps : resting Blood Pressure(mm.Hg)
5. chol : serum cholesterol in mg/dl
6. fbs : Fasting blood sugar( > 120 mg/dl)(1 = Yes, 0 = No)
6. restecg : resting electrocardiographic results. 0: normal, 1: having ST-T wave abnormality, 2: showing *probable or definite left ventricular hypertrophy by Estes' criteria
6. thalach : Maximum heart rate achieved
7. exang : exercise induced angina(1 = True, 0 = False)
8. oldpeak: ST depression induced by exercise relative to rest
9. slope : slope: the slope of the peak exercise ST segment (1: upsloping, 2: flat, 3: downsloping)
10. ca =  number of major vessels (0-3) colored by flourosopy
11. thal : 3 = normal; 6 = fixed defect; 7 = reversable defect

```python

#function to create Interactive Donut Chart
def double_donut_chart(data_1, data_2, colors, labels, explodes, title):
    colors = colors
    labels = labels
    explodes = explodes

    fig, axes = plt.subplots(1,2, figsize=(10, 6))
    
    axes[0].pie(
        data_1,
        colors=colors,
        explode=explodes,
        autopct='%1.1f%%',
    )
    axes[0].set_title("Heart Disease")



    axes[1].pie(
        data_2,
        colors=colors,
        explode=explodes,
        autopct='%1.1f%%',
    )
    axes[1].set_title("No Heart Disease")

    axes[1].legend(labels , loc='upper right', bbox_to_anchor=(1.4, 1))


    #adding Circle
    axes[0].add_artist(plt.Circle((0, 0), 0.40, fc='white'))
    axes[1].add_artist(plt.Circle((0, 0), 0.40, fc='white'))


    # Adding Title of chart
    fig.suptitle(title)

    # Displaying Chart
    plt.show()
```

```python
double_donut_chart(df_pos.groupby("sex")["target"].count(), df_neg.groupby("sex")["target"].count(), ["#E6DAA6", "#04D8B2"], ["Female", "Male"],[0.02, 0.02], "Patients having or not having Heart Disease on the basis of Sex.")
```

## Model Development

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
```

```python
categorical_features = ['sex', 'cp', 'restecg', 'slope', 'thal']
not_update = ['fbs', 'exang']
numerical_features = ["age", "trestbps", "chol", "thalach", "oldpeak", "ca"]
```

```python
#Define preprocessing Steps
numeric_transformer = Pipeline(steps=[
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("encoder", OneHotEncoder())
])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numerical_features),
    ('cat', categorical_transformer, categorical_features)
], remainder='passthrough')

preprocessor
```

```python
X = preprocessor.fit_transform(df[categorical_features+not_update+numerical_features])
#Feature Set
X[0:5]
```

```python
#target variable
y = df.target.values
y[0:5]
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=1)
print("Train Set: ", X_train.shape, y_train.shape)
print("Test Set: ", X_test.shape, y_test.shape)
```

***K_nearest Algorithm***

```python
#Classifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import ConfusionMatrixDisplay, classification_report,roc_auc_score
```

```python
parameters = {
    "n_neighbors" :list( range(1,10)),
    "leaf_size" : list(range(1,20)),
    "p" : [1,2]
}
KNN = KNeighborsClassifier()
grid_search = GridSearchCV(estimator=KNN, param_grid=parameters, cv=4)
grid_search.fit(X_train, y_train)
grid_search.best_estimator_
```

```python
y_knn_pred = grid_search.best_estimator_.predict(X_test)
ConfusionMatrixDisplay.from_predictions(y_test, y_knn_pred, cmap="GnBu")
plt.show()
```

```python
print(classification_report(y_test, y_knn_pred))
```

```python

print("ROC Score for K Neighbors Classifier is {:.1f}%" .format(round(roc_auc_score(y_test, y_knn_pred)*100)))
```

***Decision Tree***

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_parameters = {
    "criterion" : ["gini", "entropy"],
    "max_depth" : [3,5],
    "max_features" : [2, 3, 4],
    "max_leaf_nodes" : [2, 3, 4],
    "random_state" : [42]
}
decision_tree_estimator = DecisionTreeClassifier(criterion='entropy')
gs_tree = GridSearchCV(estimator=decision_tree_estimator,
                       param_grid=tree_parameters, cv=4)
gs_tree.fit(X_train, y_train)
gs_tree.best_estimator_
```

```python
y_tree_pred = gs_tree.best_estimator_.predict(X_test)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_tree_pred, cmap='GnBu')
plt.show()
```

```python
print(classification_report(y_test, y_tree_pred))
```

```python

print("ROC Score for Decision Tree Classifier is {:.1f}%" .format(round(roc_auc_score(y_test, y_tree_pred)*100)))
```

***Logistic regression***

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_param = {
    "solver" : ["lbfgs", "liblinear", "newton-cg", "newton-cholesky"],
    "penalty" : ["l2"],
    "C" : [100, 10, 1.0, 0.1, 0.01]
}
logistic_estimator = LogisticRegression()
logistic_gs = GridSearchCV(estimator=logistic_estimator, param_grid=logistic_param, cv=4)
logistic_gs.fit(X_train, y_train)
logistic_gs.best_estimator_
```

```python
logistic_gs.best_params_
```

```python
y_log_pred = logistic_gs.best_estimator_.predict(X_test)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_log_pred, cmap="GnBu")
plt.show()
```

```python

print("ROC Score for Logistic Regression Classifier is {:.1f}%" .format(round(roc_auc_score(y_test, y_log_pred)*100)))
```

***SVM***

```python

from sklearn.svm import SVC
estimator = SVC()

parameters = [{'kernel':['rbf'],
               'C':[1,10,100,1000],
               'gamma':[1,0.1,0.001,0.0001],
            },
            {'kernel':['poly'],
               'C':[1,10,100,1000],
               'gamma':[1,0.1,0.001,0.0001],
                'degree':range(1,5)
             },
             {'kernel':['linear'],
               'C':[1,10,100,1000],
            },
            {
                'kernel' : ["sigmoid"],
                'gamma':[1,0.1,0.001,0.0001],
                'C':[1,10,100,1000]
            }
        ]
```

```python
gs_svm = GridSearchCV(
    estimator=estimator,
    param_grid=parameters,
    cv=4
)

gs_svm.fit(X_train, y_train)
gs_svm.best_estimator_
```

```python
gs_svm.best_params_
```

```python
y_svm_pred = gs_svm.best_estimator_.predict(X_test)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_svm_pred, cmap="GnBu")
plt.show()
```

```python

print("ROC Score for SVM Classifier is {:.1f}%" .format(round(roc_auc_score(y_test, y_svm_pred)*100)))
```

### Conclusion

```python
from sklearn.metrics import roc_curve

#roc curve for Models
#KNN
k_fpr, k_tpr, k_thresh = roc_curve(y_test, y_knn_pred, pos_label=1)

#Decision Tree
t_fpr, t_tpr, t_thresh = roc_curve(y_test, y_tree_pred, pos_label=1)

#Logistic Regression
l_fpr, l_tpr, l_thresh = roc_curve(y_test, y_log_pred, pos_label=1)

#SVM
s_fpr, s_tpr, s_thresh = roc_curve(y_test, y_svm_pred, pos_label=1)
```
